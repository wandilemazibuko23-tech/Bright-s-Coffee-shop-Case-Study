```sql

 =========================================
-- BRIGHT COFFEE SHOP DATA ANALYSIS
-- =========================================

-- 1. View full dataset
SELECT *
FROM workspace.default.bright_coffee_shop;

-- =========================================
-- DATA EXPLORATION
-- =========================================

-- 2. Store locations
SELECT DISTINCT store_location
FROM workspace.default.bright_coffee_shop;

-- 3. Date range
SELECT 
    MIN(transaction_date) AS Minimum_Date,
    MAX(transaction_date) AS Maximum_Date
FROM workspace.default.bright_coffee_shop;

-- 4. Product categories
SELECT DISTINCT product_category
FROM workspace.default.bright_coffee_shop
ORDER BY product_category;

-- 5. Product details
SELECT DISTINCT product_detail
FROM workspace.default.bright_coffee_shop;

-- 6. Check for NULL values
SELECT *
FROM workspace.default.bright_coffee_shop
WHERE unit_price IS NULL
   OR transaction_date IS NULL
   OR transaction_time IS NULL
   OR transaction_qty IS NULL
   OR store_id IS NULL
   OR store_location IS NULL
   OR product_id IS NULL
   OR product_category IS NULL
   OR product_type IS NULL
   OR product_detail IS NULL;

-- 7. Price range
SELECT 
    MIN(unit_price) AS Minimum_price,
    MAX(unit_price) AS Maximum_price
FROM workspace.default.bright_coffee_shop;

-- =========================================
-- FEATURE ENGINEERING (ENHANCED DATASET) ---new coloumns---
-- =========================================

SELECT 
    *,
    DAY(transaction_date)   AS Day_name,
    MONTH(transaction_date) AS Month_name,
    YEAR(transaction_date)  AS Year_name,
    CASE 
        WHEN DAYNAME(transaction_date) IN ('Saturday', 'Sunday') THEN 'Weekend'
        ELSE 'Weekday'
    END AS Day_Type,
    CASE 
        WHEN HOUR(transaction_time) BETWEEN 5 AND 8 THEN '01 Morning'
        WHEN HOUR(transaction_time) BETWEEN 9 AND 11 THEN '02 Mid-day'
        WHEN HOUR(transaction_time) BETWEEN 12 AND 16 THEN '03 Afternoon'
        WHEN HOUR(transaction_time) BETWEEN 17 AND 20 THEN '04 Evening'
        ELSE '05 Night'
    END AS Time_of_Day,
    CASE 
        WHEN transaction_qty < 50 THEN 'Low Spend'
        WHEN transaction_qty BETWEEN 51 AND 200 THEN 'Medium Spend'
        WHEN transaction_qty BETWEEN 201 AND 300 THEN 'High Spend'
        ELSE 'Very High Spend'
    END AS Spend_Bucket,
    (unit_price * transaction_qty) AS Revenue
FROM workspace.default.bright_coffee_shop;

-- =========================================
-- BUSINESS INSIGHTS
-- =========================================

-- 8. Total Revenue
SELECT 
    SUM(unit_price * transaction_qty) AS Total_Revenue
FROM workspace.default.bright_coffee_shop;

-- 9. Revenue by Product Category
SELECT 
    product_category,
    SUM(unit_price * transaction_qty) AS Total_Revenue
FROM workspace.default.bright_coffee_shop
GROUP BY product_category
ORDER BY Total_Revenue DESC;

-- 10. Revenue by Time of Day
SELECT 
    CASE 
        WHEN HOUR(transaction_time) BETWEEN 5 AND 8 THEN '01 Morning'
        WHEN HOUR(transaction_time) BETWEEN 9 AND 11 THEN '02 Mid-day'
        WHEN HOUR(transaction_time) BETWEEN 12 AND 16 THEN '03 Afternoon'
        WHEN HOUR(transaction_time) BETWEEN 17 AND 20 THEN '04 Evening'
        ELSE '05 Night'
    END AS Time_of_Day,
    SUM(unit_price * transaction_qty) AS Total_Revenue
FROM workspace.default.bright_coffee_shop
GROUP BY Time_of_Day
ORDER BY Total_Revenue DESC;

-- 11. Revenue by Day Type (Weekend vs Weekday)
SELECT 
    CASE 
        WHEN DAYNAME(transaction_date) IN ('Saturday', 'Sunday') THEN 'Weekend'
        ELSE 'Weekday'
    END AS Day_Type,
    SUM(unit_price * transaction_qty) AS Total_Revenue
FROM workspace.default.bright_coffee_shop
GROUP BY Day_Type;

-- 12. Top Selling Products
SELECT 
    product_detail,
    SUM(transaction_qty) AS Total_Quantity_Sold
FROM workspace.default.bright_coffee_shop
GROUP BY product_detail
ORDER BY Total_Quantity_Sold DESC
```
